```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar)  # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar, rwidth=0.9)
```

```python
plt.hist(blood_sugar, bins=4, rwidth=0.8)
```

```python
plt.hist(blood_sugar, bins=4, rwidth=0.8, histtype='step')
```

<h3 style='color:purple'>Histogram showing normal, pre-diabetic and diabetic patients distribution</h3>
<ul style="color:brown;font-weight: bold;">
    <li>80-100: Normal</li>
    <li>100-125: Pre-diabetic</li>
    <li>80-100: Diabetic</li>
</ul>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

### Mutiple data samples in a histogram

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.94, color=['g','orange'], label=['men', 'women'])

plt.legend()
```

### horizontal orientation

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.94, color=['g','orange'], label=['men', 'women'], orientation='horizontal')

plt.legend()
```
